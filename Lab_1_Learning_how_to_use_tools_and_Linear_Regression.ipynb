아래의 `wget` 명령어를 통해서 당뇨병 데이터를 다운받겠습니다.

다운이 완료되었으면 `numpy`라이브러리를 활용해 당뇨병 데이터가 어떻게 구성되어있는지 확인해봅시다.

```
import numpy as np

xy = np.loadtxt('./diabetes.csv.gz', delimiter=',', dtype=np.float32)
print("data[0]: ", xy[0], '\n', "len: ", len(xy))
```

결과를 확인해보면 당뇨병 데이터는 8개의 관측 값(나이, 성별 등 의 정규화된 값; 여기서는 `-0.294118, 0.487437, 0.180328, -0.292929, 0., 0.00149028, -0.53117, -0.0333333`)과 당뇨병에 해당하는지의 결과 값(0 or 1)으로 구성되어있는 세트가 총 759개 포함되어 있는 것을 볼 수 있습니다.


### 학습 모델 구성

이제 아래의 코드를 따라 학습 모델을 구성하여봅시다.

우선 학습을 위한 라이브러리를 불러보겠습니다.

```
import torch
from torch.autograd import Variable
```

라이브러리를 잘 불러왔으면, 아래 명령어를 통해 8개의 입력데이터와 1개의 정답 데이터(당뇨병 여부)를 구분하여 보겠습니다.

```
x_data = Variable(torch.from_numpy(xy[:, 0:-1]))
y_data = Variable(torch.from_numpy(xy[:, [-1]]))
```

이제 위의 데이터를 학습할 수 있는 모델을 설계해보겠습니다. 아직은 코드의 모든 내용을 이해할 필요는 없습니다.

```
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.l1 = torch.nn.Linear(8, 16)  # 8개의 입력을 받아 16개의 중간 결과를 산출하는 함수
        self.l2 = torch.nn.Linear(16, 32) # 위의 16개 중간결과를 통해 16개의 입력을 받아 32개의 다음 결과를 산출하는 함수
        self.l3 = torch.nn.Linear(32, 16) # ..
        self.l4 = torch.nn.Linear(16, 1)  # 16개의 이전 결과를 통해 1개의 최종 결과를 산출하는 함수
        
        self.sigmoid = torch.nn.Sigmoid() # 최종결과를 0과 1사이의 확률 값으로 도출하는 함수
    
    def forward(self, x):                 # 위에서 정의한 함수들을 입력 x에 대해서 실행하는 함수
        out1 = self.sigmoid(self.l1(x))
        out2 = self.sigmoid(self.l2(out1))
        out3 = self.sigmoid(self.l3(out2))
        y_pred = self.sigmoid(self.l4(out3))
        
        return y_pred
```

이제 모델을 사용하기 위해 위에 정의한 모델 클래스를 선언해봅시다.

```
model = Model()
```

이제 학습을 위한 준비가 끝났습니다. 본격적으로 학습을 하기 이전의 모델 성능이 어떤지 확인해볼까요? 759개는 너무 많으니 10개만 추려서 그래프를 그려봅시다.

```
y_pred = model(x_data)            # 모델에 입력 값 (8개 입력 데이터 더미)을 입력함

t = Variable(torch.Tensor([0.5])) # 확률값이 0.5 이하는 0, 이상은 1로 변경함.
out = (y_pred > t).float() * 1    # 확률값이 0.5 이하는 0, 이상은 1로 변경함.
```

```
import matplotlib.pyplot as plt
plot_range = [i for i in range(10)] # 차트의 범위를 10개로 설정
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.scatter(plot_range, y_data.view(-1).numpy()[:10], color='b') # 정답은 파랑색
ax.scatter(plot_range, out.view(-1).detach().numpy()[:10], color='r') # 예측은 빨간색
```

첫 모델은 랜덤한 숫자들로 초기화되어있기 때문에 위의 결과는 다르겠지만, 여기서는 10개중에 6개를 맞춘 것으로 보이네요. 이번엔 간단한 함수로 759개의 전체 퍼센트를 확인해볼까요?

```
def calcResult(preds, y_data):
  total_samples = len(y_data)
  correct_samples = 0
  for pred, y in zip(preds, y_data):
    if pred == y: 
      correct_samples += 1
  print(correct_samples/total_samples * 100, "%")
```

```
calcResult(out, y_data)
```

이번 실험에서는 **34%**의 수치가 나왔네요, 이 결과는 위의 설명처럼 매번 달라질 수는 있지만 34%는 병원에서 사용하기에는 턱없이 부족한 수치네요..

이제 이 결과를 개선해봅시다!

```
criterion = torch.nn.MSELoss(reduction='mean')
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

학습을 Gradient Descent알고리즘과 Mean Square Error를 통해 학습을 해봅시다.

*(아마, 위 내용들은 이어지는 다음 수업에서 자세하게 다룰 예정입니다. 이번 실습은 따라 돌려보고 잘 되는 것만 우선 확인해보자고요~~)*

이제 학습을 해봅시다. 음 적당히 많은 횟수의 학습을 하면 좋겠는데, 100번을 해볼까요?

```
for epoch in range(100):
    y_pred = model(x_data) # 모델에 입력 값을 넣어 예측 값을 산출함

    loss = criterion(y_pred, y_data) # 산출 결과와 정답을 비교하여 차이를 계산함 (이어지는 수업에서 더 자세하게 공부해봅시다)
    print(f'Epoch: {epoch + 1}/100000 | Loss: {loss.item():.4f}')

    optimizer.zero_grad()
    loss.backward() # 차이가 줄어드는 방향으로 모델을 수정하는 함수
l    optimizer.step()
```

정답과의 차이(Loss)가 줄기는 하는데 뭔가 시원찮아 보이네요. 그래도 한번 성능을 볼까요?

```
t = Variable(torch.Tensor([0.5]))
out = (y_pred > t).float() * 1

calcResult(out, y_data)
```

오, 그래도 나름 **30%이상 향상**된 결과가 나왔네요, 여기서 멈추지말고 더 많이 학습해보죠 **100,000번**이면 누가봐도 많이 학습하는 것으로 보이겠네요. 

```
for epoch in range(100000):
    y_pred = model(x_data) # 모델에 입력 값을 넣어 예측 값을 산출함

    loss = criterion(y_pred, y_data) # 산출 결과와 정답을 비교하여 차이를 계산함 (이어지는 수업에서 더 자세하게 공부해봅시다)
    if epoch % 1000 == 0: # 100,000번을 화면에 띄우면 너무 많으니, 1,000번당 한번씩만 보자고요.
      print(f'Epoch: {epoch + 1}/100000 | Loss: {loss.item():.4f}')

    optimizer.zero_grad()
    loss.backward() # 차이가 줄어드는 방향으로 모델을 수정하는 함수
    optimizer.step()
```

이번 실험은 확실히 Loss가 많이 줄은 것이 보이네요. 바로 결과를 확인해보죠!

```
t = Variable(torch.Tensor([0.5]))
out = (y_pred > t).float() * 1

calcResult(out, y_data)
```

오호.. 거의 **80% 정도** 되는 성능이 나왔어요.
